```python
## Basic S3 Operations ##
```

```python
# 1. List all S3 buckets in your account
aws s3 ls

# 2. Create a new S3 bucket (replace with unique name)
aws s3 mb s3://sab-my-student-bucket-12345

# 3. Upload a file to S3 (Make a file sample.txt first)
aws s3 cp sample.txt s3://sab-my-student-bucket-12345/

# 4. List contents of a specific bucket
aws s3 ls s3://sab-my-student-bucket-12345/

# 5. Download a file from S3
aws s3 cp s3://sab-my-student-bucket-12345/sample.txt ./downloaded.txt

### Now you can review these resources in your AWS Console if you want ###
### Next we will clean up the resources created ###

# 6. Delete a specific file from S3
aws s3 rm s3://sab-my-student-bucket-12345/sample.txt

# 7. Delete the empty bucket
aws s3 rb s3://sab-my-student-bucket-12345
```

```python
## IAM and Security Commands ##
```

```python
# 6. List IAM users in the account
aws iam list-users --output table

# 7. Get details about your current IAM identity
aws sts get-caller-identity

# 8. List all IAM roles (useful for understanding permissions)
aws iam list-roles --query 'Roles[*].[RoleName,Arn]' --output table
```

```python
## EC2 and Compute commands ##
```

```python
# 9. List all EC2 instances with their states
aws ec2 describe-instances --query 'Reservations[*].Instances[*].[InstanceId,State.Name,InstanceType]' --output table

# 10. List all available EC2 instance types in a region
aws ec2 describe-instance-types --query 'InstanceTypes[*].[InstanceType,VCpuInfo.DefaultVCpus,MemoryInfo.SizeInMiB]' --output table --max-items 10

# 11. Describe available regions
aws ec2 describe-regions --query 'Regions[*].[RegionName,Endpoint]' --output table
```

```python
## Log and Monitoring commands ##
```

```python
# 12. List CloudWatch log groups
aws logs describe-log-groups --query 'logGroups[*].[logGroupName,storedBytes]' --output table

# 13. Get recent CloudWatch log events (replace log-group-name)
aws logs tail /aws/lambda/my-function --follow
```

```python
## Advanced Commands ##
```

```python
# 14. List Lambda functions in your account
aws lambda list-functions --query 'Functions[*].[FunctionName,Runtime,LastModified]' --output table

# 15. Create a JSON output of S3 bucket properties (demonstrates JSON processing)
aws s3api list-buckets --query 'Buckets[*].[Name,CreationDate]' --output json > my-buckets.json
```
