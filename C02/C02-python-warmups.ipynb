```python
# ============================================================================
# Exercise 1: Object-Oriented Programming (Classes and Objects)
# ============================================================================
print("Exercise 1: Object-Oriented Programming")
print("-" * 40)

class Dog:
    """A simple Dog class demonstrating OOP concepts"""
    
    # Class variable (shared by all instances)
    species = "Canis familiaris"
    
    def __init__(self, name, age, breed):
        # Instance variables (unique to each instance)
        self.name = name
        self.age = age
        self.breed = breed
    
    def bark(self):
        return f"{self.name} says Woof!"
    
    def description(self):
        return f"{self.name} is a {self.age}-year-old {self.breed}"
    
    def __str__(self):
        return f"Dog({self.name}, {self.age})"

# Create instances
buddy = Dog("Buddy", 3, "Golden Retriever")
luna = Dog("Luna", 5, "Husky")

print(buddy.description())
print(luna.bark())
print(f"Species: {Dog.species}")
print(f"String representation: {buddy}")
print()
```

```python
# ============================================================================
# Exercise 2: Yield Statement (Generators)
# ============================================================================
print("Exercise 2: Yield Statement (Generators)")
print("-" * 40)

def fibonacci(n):
    """Generator function that yields Fibonacci numbers"""
    a, b = 0, 1
    count = 0
    while count < n:
        yield a
        a, b = b, a + b
        count += 1

print("First 10 Fibonacci numbers:")
for num in fibonacci(10):
    print(num, end=" ")
print("\n")

# Generator expression
squares_gen = (x**2 for x in range(5))
print("Squares from generator expression:")
print(list(squares_gen))
print()
```

```python
# ============================================================================
# Exercise 3: Try-Except Blocks (Exception Handling)
# ============================================================================
print("Exercise 3: Try-Except Blocks")
print("-" * 40)

def divide_numbers(a, b):
    """Safely divide two numbers with error handling"""
    try:
        result = a / b
        return f"{a} / {b} = {result}"
    except ZeroDivisionError:
        return "Error: Cannot divide by zero!"
    except TypeError:
        return "Error: Invalid type - both inputs must be numbers"
    finally:
        print(f"  Attempted division of {a} by {b}")

print(divide_numbers(10, 2))
print(divide_numbers(10, 0))
print(divide_numbers("10", 2))

# Multiple exceptions
try:
    values = [1, 2, 3]
    print(f"Value at index 5: {values[5]}")
except (IndexError, KeyError) as e:
    print(f"Error: {type(e).__name__} - {e}")
print()
```

```python
# ============================================================================
# Exercise 7: Match Statement (Python 3.10+)
# ============================================================================
print("Exercise 7: Match Statement (Structural Pattern Matching)")
print("-" * 40)

def process_command(command):
    """Process commands using match statement"""
    match command.split():
        case ["quit"]:
            return "Goodbye!"
        case ["help"]:
            return "Available commands: quit, help, move [direction]"
        case ["move", direction] if direction in ["north", "south", "east", "west"]:
            return f"Moving {direction}"
        case ["move", _]:
            return "Invalid direction"
        case _:
            return "Unknown command"

commands = ["help", "move north", "move up", "quit", "jump"]
for cmd in commands:
    print(f"'{cmd}' -> {process_command(cmd)}")
print()
```

```python
# ============================================================================
# Exercise 8: Match with More Complex Patterns
# ============================================================================
print("Exercise 8: Advanced Match Patterns")
print("-" * 40)

def analyze_point(point):
    """Analyze points using pattern matching"""
    match point:
        case (0, 0):
            return "Origin"
        case (0, y):
            return f"On Y-axis at y={y}"
        case (x, 0):
            return f"On X-axis at x={x}"
        case (x, y) if x == y:
            return f"On diagonal at ({x}, {y})"
        case (x, y):
            return f"Point at ({x}, {y})"

points = [(0, 0), (0, 5), (3, 0), (4, 4), (2, 7)]
for pt in points:
    print(f"{pt} -> {analyze_point(pt)}")
print()
```

```python
# ============================================================================
# Exercise 10: Iterators
# ============================================================================
print("Exercise 10: Iterators")
print("-" * 40)

class Countdown:
    """Custom iterator that counts down"""
    def __init__(self, start):
        self.current = start
    
    def __iter__(self):
        return self
    
    def __next__(self):
        if self.current <= 0:
            raise StopIteration
        self.current -= 1
        return self.current + 1

print("Countdown from 5:")
for num in Countdown(5):
    print(num, end=" ")
print("\n")

# Using iter() and next()
my_list = ['a', 'b', 'c']
iterator = iter(my_list)
print(f"First: {next(iterator)}")
print(f"Second: {next(iterator)}")
print(f"Third: {next(iterator)}")
print()
```

```python
# ============================================================================
# Exercise 11: Regular Expressions
# ============================================================================
print("Exercise 11: Regular Expressions")
print("-" * 40)

import re

text = "Contact us at support@example.com or sales@example.com"

# Find all email addresses
emails = re.findall(r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b', text)
print(f"Emails found: {emails}")

# Pattern matching
phone = "My phone is 555-123-4567"
match = re.search(r'(\d{3})-(\d{3})-(\d{4})', phone)
if match:
    print(f"Phone: {match.group(0)}")
    print(f"Area code: {match.group(1)}")

# Replace patterns
censored = re.sub(r'\b\d{3}-\d{3}-\d{4}\b', 'XXX-XXX-XXXX', phone)
print(f"Censored: {censored}")

# Split on patterns
data = "apple,banana;cherry orange"
fruits = re.split(r'[,;\s]+', data)
print(f"Split fruits: {fruits}")
print()
```

```python
# ============================================================================
# Exercise 13: None Type
# ============================================================================
print("Exercise 13: None Type")
print("-" * 40)

def find_item(items, target):
    """Return item if found, None otherwise"""
    for item in items:
        if item == target:
            return item
    return None

result = find_item([1, 2, 3, 4], 3)
print(f"Found 3: {result}")

result = find_item([1, 2, 3, 4], 10)
print(f"Found 10: {result}")

# Checking for None
if result is None:
    print("Item not found")

# None vs empty vs zero
values = [None, 0, "", [], False]
for val in values:
    print(f"{repr(val):10} -> is None: {val is None}, is falsy: {not val}")

# Function with no return implicitly returns None
def no_return():
    pass

print(f"Function with no return: {no_return()}")
print()
```
