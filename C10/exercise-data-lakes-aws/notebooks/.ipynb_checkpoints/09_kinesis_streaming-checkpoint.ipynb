## Background: Batch vs. Streaming Ingestion

Every exercise so far has used **batch ingestion**: data already exists somewhere (NCBI FTP, a public S3 bucket, a local file) and we copy it into our lake in one shot. Batch works well when:

- Data arrives on a predictable schedule (nightly FASTQ uploads, weekly ClinVar releases)
- Latency of hours or days is acceptable
- Files are large and self-contained

But genomics and clinical workflows increasingly generate **continuous event streams** where batch is insufficient:

| Scenario | Why streaming? |
|---|---|
| Oxford Nanopore sequencer producing reads in real time | QC must flag low-yield runs before they finish, saving reagent cost |
| Clinical lab reporting ACMG variant classifications | Oncologist dashboard must reflect reclassifications within minutes |
| IoT biosensors in a -80°C freezer farm | Temperature excursions must trigger alerts in seconds, not the next morning |
| Patient portal: new lab results posted | EHR integration must push notifications immediately |
| Sequencer instrument logs | Real-time monitoring catches hardware faults before a run is ruined |

**Streaming ingestion** means events are ingested, processed, and (optionally) stored as they arrive — latency measured in seconds or milliseconds rather than hours.

## The AWS Kinesis Family

AWS offers three distinct Kinesis services. They are often confused — here is a clear comparison:

```
┌─────────────────────────────────────────────────────────────────────────────┐
│                         AWS Kinesis Family                                  │
│                                                                             │
│  Kinesis Data Streams          Kinesis Data Firehose                        │
│  ─────────────────────         ──────────────────────                       │
│  Low-latency pub/sub           Managed delivery to S3,                      │
│  stream (ms latency)           Redshift, OpenSearch, HTTP                   │
│  You manage consumers          Fully managed (no consumers)                 │
│  Replay up to 7 days           Buffer → compress → land                     │
│  Shard-based scaling           Auto-scaling                                 │
│                                                                             │
│  Kinesis Data Analytics                                                     │
│  ──────────────────────                                                     │
│  Real-time SQL or Flink        ◄── reads from Data Streams                 │
│  on streaming data             ◄── writes enriched stream back              │
│  Windowed aggregations                                                      │
│  Anomaly detection                                                          │
└─────────────────────────────────────────────────────────────────────────────┘
```

### The Typical Flow

```
Producers                Kinesis               Consumers / Storage
─────────                ───────               ───────────────────
Sequencer logs   ──►  Data Streams  ──►  Data Analytics  ──►  Alerts (SNS)
Variant callers  ──►  (shards)      ──►  Lambda          ──►  Database
Sensors          ──►               └──►  Firehose        ──►  S3 (bronze)
EHR events                                                ──►  Redshift
```

In the context of our data lake: **Kinesis Firehose is the canonical bridge from a live event stream into the Bronze layer of S3.**

```python
import boto3
import json
import time
import uuid
import datetime
import random

# ── CONFIGURE ────────────────────────────────────────────
STUDENT_ID  = "sab032226"      # replace with your ID
AWS_REGION  = "us-east-1"
BUCKET      = f"binfx410-datalake-{STUDENT_ID}"

STREAM_NAME   = f"binfx410-variants-{STUDENT_ID}"    # Kinesis Data Streams
FIREHOSE_NAME = f"binfx410-firehose-{STUDENT_ID}"    # Kinesis Firehose

# IAM role that Firehose assumes to write to S3
IAM_ROLE_ARN = f"arn:aws:iam::123456789012:role/binfx410-firehose-role"

# Boto3 clients
kinesis  = boto3.client('kinesis',   region_name=AWS_REGION)
firehose = boto3.client('firehose',  region_name=AWS_REGION)
s3       = boto3.client('s3',        region_name=AWS_REGION)

print(f"Stream name:   {STREAM_NAME}")
print(f"Firehose name: {FIREHOSE_NAME}")
print(f"S3 target:     s3://{BUCKET}/bronze/streaming-variants/")
```

## Section 2: Kinesis Data Streams — Creating a Stream

### Key Concepts

**Shards** are the core capacity unit of Kinesis Data Streams:
- Each shard supports **1 MB/s ingest** and **2 MB/s read**
- A stream starts with 1–N shards; you can split or merge shards to scale
- Records are distributed across shards by a **partition key** (a string you supply with each record)
- Records within a shard are **strictly ordered** by arrival time
- Records are **retained for 24 hours** (default) up to **7 days** — enabling replay

For our genomics use case:
- **Partition key = chromosome** keeps variants on the same chromosome ordered in the same shard
- 1 shard is sufficient for a single sequencer producing ~100 variant events/second
- A clinical lab processing 500 simultaneous samples might need 5–10 shards

**What this does:** This would call `kinesis.create_stream()` to provision a Kinesis Data Stream named `binfx410-variants-{STUDENT_ID}` with 1 shard in `us-east-1`. Stream creation is asynchronous — the `waiter` loop polls `DescribeStreamSummary` every 10 seconds until the stream status transitions from `CREATING` to `ACTIVE` (typically 10–30 seconds). Once active, the stream is ready to accept records.

```python
# Create a Kinesis Data Stream with 1 shard
kinesis.create_stream(
    StreamName=STREAM_NAME,
    ShardCount=1
)
print(f"Creating stream '{STREAM_NAME}'...")

# Wait for stream to become ACTIVE
waiter = kinesis.get_waiter('stream_exists')
waiter.wait(StreamName=STREAM_NAME)
print(f"Stream '{STREAM_NAME}' is ACTIVE.")

# Describe the stream to inspect its properties
desc = kinesis.describe_stream_summary(StreamName=STREAM_NAME)
summary = desc['StreamDescriptionSummary']
print(f"  ARN:    {summary['StreamARN']}")
print(f"  Shards: {summary['OpenShardCount']}")
print(f"  Retention: {summary['RetentionPeriodHours']} hours")
```

### Producing Records: The Variant Event Schema

Before writing a producer, we need to define the event schema. For a streaming variant pipeline, each record represents a single called variant with enough context for downstream consumers:

```json
{
  "event_id":      "a3f2c1d0-...",
  "timestamp":     "2025-03-15T14:32:07.123Z",
  "sample_id":     "NA12878",
  "run_id":        "RUN_20250315_001",
  "chrom":         "chr17",
  "pos":            41223094,
  "ref":           "A",
  "alt":           "T",
  "qual":           98.4,
  "depth":          52,
  "allele_freq":    0.48,
  "variant_type":  "SNP",
  "gene":          "BRCA1",
  "classification": "VUS"
}
```

**Why JSON?** Kinesis records are arbitrary bytes (up to 1 MB each). JSON is human-readable, schema-flexible, and natively parsed by Lambda, Firehose, and Kinesis Data Analytics. For very high throughput (>10,000 events/sec) you would switch to Avro or Protocol Buffers for smaller payloads.

**What this does:** This defines a `generate_variant_event()` function that returns a synthetic variant JSON record, and a `simulate_sequencer()` function that calls `kinesis.put_record()` in a loop — simulating a sequencer continuously emitting called variants as it processes reads. The partition key is the chromosome name, so all chr17 events land on the same shard (preserving positional order). Each `put_record` call is a synchronous HTTP request to the Kinesis API; at 5 records/sec this would cost roughly $0.015/hour at standard Kinesis pricing.

```python
GENES_ON_CHR17 = ['BRCA1', 'TP53', 'ERBB2', 'CDH1', 'MAP2K4']
CLASSIFICATIONS = ['Benign', 'Likely_Benign', 'VUS', 'Likely_Pathogenic', 'Pathogenic']
VARIANT_TYPES   = ['SNP', 'SNP', 'SNP', 'INDEL', 'SNP']  # SNPs are more common

def generate_variant_event(sample_id: str, run_id: str) -> dict:
    """Generate a realistic synthetic variant event."""
    chrom = random.choice(['chr17', 'chr13', 'chr22', 'chr1'])
    return {
        'event_id':      str(uuid.uuid4()),
        'timestamp':     datetime.datetime.utcnow().isoformat() + 'Z',
        'sample_id':     sample_id,
        'run_id':        run_id,
        'chrom':         chrom,
        'pos':           random.randint(1_000_000, 50_000_000),
        'ref':           random.choice(['A', 'C', 'G', 'T']),
        'alt':           random.choice(['A', 'C', 'G', 'T']),
        'qual':          round(random.uniform(20, 99), 1),
        'depth':         random.randint(10, 200),
        'allele_freq':   round(random.uniform(0.1, 0.9), 3),
        'variant_type':  random.choice(VARIANT_TYPES),
        'gene':          random.choice(GENES_ON_CHR17) if chrom == 'chr17' else 'INTERGENIC',
        'classification': random.choice(CLASSIFICATIONS),
    }

def simulate_sequencer(n_records: int = 20, delay_sec: float = 0.5):
    """Simulate a sequencer producing variant events in real time."""
    run_id = f"RUN_{datetime.date.today().strftime('%Y%m%d')}_001"
    print(f"Simulating run {run_id} — sending {n_records} variant events...")

    for i in range(n_records):
        event = generate_variant_event(sample_id='NA12878', run_id=run_id)
        response = kinesis.put_record(
            StreamName=STREAM_NAME,
            Data=json.dumps(event).encode('utf-8'),
            PartitionKey=event['chrom']          # routes by chromosome
        )
        print(f"  [{i+1:02d}] shard={response['ShardId']}  "
              f"chrom={event['chrom']}  gene={event['gene']}  "
              f"class={event['classification']}")
        time.sleep(delay_sec)

    print(f"\nDone. {n_records} records written to '{STREAM_NAME}'.")

simulate_sequencer(n_records=20, delay_sec=0.3)
```

**What this does:** This would demonstrate the **consumer** side of Kinesis Data Streams. `get_shard_iterator()` returns a pointer to a position in the shard (here `TRIM_HORIZON` = the oldest available record). `get_records()` then pulls up to 100 records from that position in a single API call, returning JSON-decoded variant events. This is the **pull model** — consumers explicitly request batches of records. AWS Lambda and the Kinesis Consumer Library (KCL) abstract this loop automatically.

```python
# Read records back from the stream (consumer side)
# First, get the shard IDs
stream_desc = kinesis.describe_stream(StreamName=STREAM_NAME)
shard_id = stream_desc['StreamDescription']['Shards'][0]['ShardId']

# Get a shard iterator starting at the oldest available record
iterator_resp = kinesis.get_shard_iterator(
    StreamName=STREAM_NAME,
    ShardId=shard_id,
    ShardIteratorType='TRIM_HORIZON'   # start from oldest record
)
shard_iterator = iterator_resp['ShardIterator']

# Pull records from the shard
records_resp = kinesis.get_records(ShardIterator=shard_iterator, Limit=100)
records = records_resp['Records']

print(f"Retrieved {len(records)} records from shard {shard_id}:")
for rec in records[:5]:   # show first 5
    event = json.loads(rec['Data'])
    print(f"  seq={rec['SequenceNumber'][:20]}...  "
          f"chrom={event['chrom']}  gene={event['gene']}  "
          f"class={event['classification']}")
print(f"  ... ({len(records) - 5} more records)")
```

## Section 3: Kinesis Data Firehose — Streaming to S3

Kinesis Data Streams requires you to write and operate your own consumer. If your goal is simply **landing records to S3** (or Redshift, or OpenSearch), **Kinesis Data Firehose** handles that automatically.

### How Firehose Works

```
Producer
  │
  │  put_record() or from Kinesis Data Streams
  ▼
┌─────────────────────────────────────────────┐
│         Kinesis Data Firehose               │
│                                             │
│  Buffer window: 60 sec  OR  5 MB            │
│  (whichever comes first)                    │
│                                             │
│  Optional transforms:                       │
│   • Lambda (enrich, filter, reformat)       │
│   • Format conversion (JSON → Parquet)      │
│   • Compression (GZIP, Snappy, ZSTD)        │
└───────────────┬─────────────────────────────┘
                │
                ▼  partitioned prefix
  s3://bucket/bronze/streaming-variants/
              year=2025/month=03/day=15/hour=14/
                  binfx410-firehose-2025-03-15-14-32-07-abc123.gz
```

**Key difference from Data Streams:**
- No shard management — Firehose scales automatically
- No consumer code to write — Firehose IS the consumer
- Minimum latency is the buffer window (60 seconds by default)
- You give up sub-second latency in exchange for operational simplicity

For a data lake use case (landing to Bronze S3), Firehose is almost always the right choice over writing your own Kinesis consumer.

**What this does:** This would call `firehose.create_delivery_stream()` to create a Firehose delivery stream configured to buffer records for up to 60 seconds or 5 MB (whichever comes first), compress them with GZIP, and write them to `s3://binfx410-datalake-{STUDENT_ID}/bronze/streaming-variants/` with dynamic partitioning using the `!{timestamp:yyyy/MM/dd/HH}` prefix pattern. The `!{firehose:random-string}` suffix prevents collisions when multiple Firehose shards write to the same prefix simultaneously.

```python
firehose.create_delivery_stream(
    DeliveryStreamName=FIREHOSE_NAME,
    DeliveryStreamType='DirectPut',          # producers call Firehose directly
    ExtendedS3DestinationConfiguration={
        'RoleARN': IAM_ROLE_ARN,
        'BucketARN': f'arn:aws:s3:::{BUCKET}',
        'Prefix': (
            'bronze/streaming-variants/'
            'year=!{timestamp:yyyy}/'
            'month=!{timestamp:MM}/'
            'day=!{timestamp:dd}/'
            'hour=!{timestamp:HH}/'
        ),
        'ErrorOutputPrefix': 'bronze/streaming-variants-errors/',
        'BufferingHints': {
            'SizeInMBs': 5,          # flush when buffer reaches 5 MB
            'IntervalInSeconds': 60  # OR after 60 seconds (whichever first)
        },
        'CompressionFormat': 'GZIP',
    }
)

# Wait for Firehose to become ACTIVE
import time
while True:
    resp = firehose.describe_delivery_stream(DeliveryStreamName=FIREHOSE_NAME)
    status = resp['DeliveryStreamDescription']['DeliveryStreamStatus']
    if status == 'ACTIVE':
        print(f"Firehose '{FIREHOSE_NAME}' is ACTIVE.")
        break
    print(f"  Status: {status} — waiting...")
    time.sleep(5)
```

**What this does:** This would call `firehose.put_record_batch()` to send 50 synthetic variant events to the Firehose delivery stream in batches of 10 (the API maximum per batch call is 500, but smaller batches improve error isolation). Each record is JSON-encoded and newline-delimited — the newline is important because Firehose concatenates records in the buffer and the newline allows downstream tools to split lines correctly. After the batch is sent, the records sit in Firehose's buffer until the 60-second window or 5 MB threshold is reached, at which point Firehose compresses and writes a GZIP file to S3.

```python
def send_events_to_firehose(n_events: int = 50, batch_size: int = 10):
    """Send synthetic variant events to Firehose in batches."""
    run_id = f"RUN_{datetime.date.today().strftime('%Y%m%d')}_002"
    all_events = [
        generate_variant_event('NA12878', run_id) for _ in range(n_events)
    ]

    total_sent = 0
    for i in range(0, n_events, batch_size):
        batch = all_events[i:i + batch_size]
        records = [
            # Newline-delimited JSON — critical for correct S3 file splitting
            {'Data': (json.dumps(evt) + '\n').encode('utf-8')}
            for evt in batch
        ]
        resp = firehose.put_record_batch(
            DeliveryStreamName=FIREHOSE_NAME,
            Records=records
        )
        failed = resp.get('FailedPutCount', 0)
        total_sent += len(batch) - failed
        print(f"  Batch {i//batch_size + 1}: sent {len(batch) - failed}/{len(batch)} records")

    print(f"\nTotal sent: {total_sent}/{n_events}")
    print(f"Records are buffering in Firehose — will land in S3 within ~60 seconds.")
    print(f"S3 prefix: s3://{BUCKET}/bronze/streaming-variants/year=.../month=.../")

send_events_to_firehose(n_events=50)
```

**What this does:** This would wait 75 seconds for the Firehose buffer to flush, then call `S3.list_objects_v2` on the `bronze/streaming-variants/` prefix to verify that GZIP files have landed. It then reads one of the GZIP files using `S3.get_object` and `gzip.decompress()` to confirm the JSON records are intact. This verification step demonstrates the end-to-end path: producer → Firehose → S3 bronze zone.

```python
import gzip
import io

print("Waiting 75 seconds for Firehose buffer to flush to S3...")
time.sleep(75)

# Check what landed in the bronze zone
response = s3.list_objects_v2(
    Bucket=BUCKET,
    Prefix='bronze/streaming-variants/'
)
objects = response.get('Contents', [])
print(f"Files landed in S3: {len(objects)}")
for obj in objects:
    print(f"  {obj['Key']}  ({obj['Size']} bytes)")

if objects:
    # Read the first file and show a sample record
    first_key = objects[0]['Key']
    body = s3.get_object(Bucket=BUCKET, Key=first_key)['Body'].read()
    text = gzip.decompress(body).decode('utf-8')
    lines = [l for l in text.strip().split('\n') if l]
    print(f"\nSample record from {first_key}:")
    print(json.dumps(json.loads(lines[0]), indent=2))
    print(f"  ... ({len(lines)} total records in this file)")
```

## Section 4: Kinesis Data Analytics — Real-Time SQL on Streams

Kinesis Data Analytics lets you run SQL (or Apache Flink) **directly on the stream** — before records ever land in S3. This enables:

- **Windowed aggregations:** "How many Pathogenic variants have been called in the last 60 seconds?"
- **Anomaly detection:** "Is the allele frequency distribution for this sample unusual?"
- **Real-time filtering:** "Route only BRCA1/BRCA2 variants to an urgent-review stream"
- **Running counts and thresholds:** Trigger an SNS alert when TMB > 10 mut/Mb

### Tumbling vs. Sliding Windows

```
Tumbling window (non-overlapping):          Sliding window (overlapping):
─────────────────────────────────           ────────────────────────────
[--60s--][--60s--][--60s--]...              [--60s--]
         Aggregate per window               [--60s--]
                                                     [--60s--]
  Good for: per-minute dashboards             Good for: rolling averages,
  counters, billing summaries                 smoothed QC metrics
```

Kinesis Data Analytics SQL uses the `ROWTIME` pseudo-column (the time the record entered the stream) for window boundaries.

**What this does:** This defines the SQL application schema and query that would be deployed to Kinesis Data Analytics. The `CREATE STREAM` statement defines the input schema matching our variant JSON events. The `CREATE PUMP` continuously reads from the input stream and writes aggregated rows to `DESTINATION_STREAM`. Every 60 seconds (tumbling window), the query counts variants by classification and chromosome, computing average quality — producing a real-time dashboard feed that can be wired to CloudWatch, a clinical alerting system, or a downstream Firehose for archival.

```python
# This SQL would be deployed to a Kinesis Data Analytics application
# It runs continuously against the live stream — not against stored data

ANALYTICS_SQL = """
-- 1. Define the input stream schema
CREATE OR REPLACE STREAM "SOURCE_SQL_STREAM_001" (
    event_id        VARCHAR(64),
    sample_id       VARCHAR(32),
    chrom           VARCHAR(8),
    pos             INTEGER,
    gene            VARCHAR(32),
    qual            DOUBLE,
    variant_type    VARCHAR(16),
    classification  VARCHAR(32)
);

-- 2. Output stream: per-minute variant classification counts
CREATE OR REPLACE STREAM "VARIANT_SUMMARY_STREAM" (
    window_end      TIMESTAMP,
    chrom           VARCHAR(8),
    classification  VARCHAR(32),
    variant_count   INTEGER,
    avg_qual        DOUBLE
);

-- 3. Pump: runs the windowed aggregation continuously
CREATE OR REPLACE PUMP "VARIANT_SUMMARY_PUMP" AS
INSERT INTO "VARIANT_SUMMARY_STREAM"
SELECT STREAM
    STEP("SOURCE_SQL_STREAM_001".ROWTIME BY INTERVAL '60' SECOND) AS window_end,
    chrom,
    classification,
    COUNT(*)             AS variant_count,
    AVG(qual)            AS avg_qual
FROM "SOURCE_SQL_STREAM_001"
GROUP BY
    STEP("SOURCE_SQL_STREAM_001".ROWTIME BY INTERVAL '60' SECOND),
    chrom,
    classification;
"""

print("Analytics SQL defined. This would be deployed via:")
print("  kinesis_analytics.create_application(")
print(f"    ApplicationName='binfx410-variant-analytics-{STUDENT_ID}',")
print("    Inputs=[{'KinesisStreamsInput': {'ResourceARN': stream_arn}}],")
print("    ApplicationCode=ANALYTICS_SQL")
print("  )")
```

**What this does:** This defines a `check_for_pathogenic_alerts()` function that would be invoked by a Lambda function connected to the `VARIANT_SUMMARY_STREAM` output. It parses each aggregated window record and — if any `Pathogenic` variants are detected — publishes a notification to an SNS topic using `sns.publish()`. In a clinical deployment, this SNS topic would be subscribed to by the oncology team's paging system, an EHR integration bus, or a Slack webhook, enabling near-real-time clinical alerting from sequencing data.

```python
# This function would run inside a Lambda triggered by the analytics output stream
# Lambda receives batches of records from the VARIANT_SUMMARY_STREAM

import boto3

SNS_TOPIC_ARN = "arn:aws:sns:us-east-1:123456789012:binfx410-clinical-alerts"

def check_for_pathogenic_alerts(kinesis_records: list) -> None:
    """Lambda handler: fire SNS alert if Pathogenic variants detected."""
    sns = boto3.client('sns')

    for record in kinesis_records:
        # Kinesis records are base64-encoded in Lambda event
        import base64
        payload = json.loads(base64.b64decode(record['kinesis']['data']))

        if payload['classification'] == 'Pathogenic' and payload['variant_count'] > 0:
            message = (
                f"ALERT: {payload['variant_count']} Pathogenic variant(s) detected\n"
                f"Gene region: {payload['chrom']}\n"
                f"Window ending: {payload['window_end']}\n"
                f"Average quality: {payload['avg_qual']:.1f}\n"
                f"Action: Review in variant interpretation portal"
            )
            sns.publish(
                TopicArn=SNS_TOPIC_ARN,
                Subject="[BINFX410] Pathogenic Variant Alert",
                Message=message
            )
            print(f"Alert sent for {payload['variant_count']} Pathogenic variants on {payload['chrom']}")

print("Alert Lambda defined. Would be deployed and triggered by Kinesis Data Analytics output.")
```

## Section 5: Full Streaming Pipeline Architecture

Putting all three Kinesis services together with the data lake from previous exercises:

```
┌────────────────────────────────────────────────────────────────────────────┐
│                  BINFX410 Streaming Genomics Pipeline                      │
│                                                                            │
│  PRODUCERS                                                                 │
│  ─────────                                                                 │
│  Sequencer      ──put_record──►  Kinesis Data Streams                     │
│  Variant caller                  (1 shard, chr partition key)              │
│  EHR events                         │                                     │
│                                      ├──► Kinesis Data Analytics          │
│                                      │     (60s tumbling window SQL)      │
│                                      │        │                           │
│                                      │        ├──► PATHOGENIC → SNS alert │
│                                      │        └──► stats → CloudWatch     │
│                                      │                                    │
│                                      └──► Kinesis Firehose                │
│                                            (60s buffer, GZIP)             │
│                                               │                           │
│  DATA LAKE                                    ▼                           │
│  ─────────                         S3 bronze/streaming-variants/          │
│  Bronze ◄── raw GZIP files            year=2025/month=03/day=15/         │
│  Silver ◄── Glue ETL (hourly)             *.json.gz                       │
│  Gold   ◄── Athena CTAS (daily)                                           │
└────────────────────────────────────────────────────────────────────────────┘
```

### Integration with the Medallion Architecture

| Layer | Source | Schedule | Format |
|---|---|---|---|
| **Bronze** | Firehose → S3 directly | Real-time (60s lag) | GZIP JSON, Hive partitioned by date |
| **Silver** | Glue ETL reads bronze | Hourly trigger | Parquet, partitioned by chrom |
| **Gold** | Athena CTAS from silver | Daily | Parquet, aggregated per gene/run |

The streaming bronze layer is structurally identical to the batch bronze layer from Exercise 1 — Athena and Glue don't care how the files got there.

## Section 6: Kinesis Cost Model

Understanding Kinesis pricing is essential for designing production genomics pipelines.

### Kinesis Data Streams Pricing (us-east-1)

| Component | Price |
|---|---|
| Shard hour | $0.015 / shard / hour |
| PUT payload unit (25 KB) | $0.014 per million units |
| Extended data retention (7 days) | $0.020 / shard / hour extra |

**Example:** 1 shard running 24/7 for a month = 1 × 24 × 30 × $0.015 = **$10.80/month**  
Plus PUT costs: 10,000 events/day × 30 days × (1 KB / 25 KB) × $0.014/million ≈ **$0.0002/month** (negligible)

### Kinesis Data Firehose Pricing

| Component | Price |
|---|---|
| Data ingested | $0.029 per GB |
| Format conversion (JSON → Parquet) | $0.018 per GB |

**Example:** 10 GB/day of variant JSON = 10 × 30 × $0.029 = **$8.70/month**  
After GZIP compression (typical 5x ratio on JSON): effective cost ≈ **$1.74/month**

### Kinesis Data Analytics Pricing

| Component | Price |
|---|---|
| Kinesis Processing Unit (KPU) | $0.11 per KPU-hour |
| 1 KPU = 1 vCPU + 4 GB RAM | |

**Example:** 1 KPU running 24/7 = 24 × 30 × $0.11 = **$79.20/month**  
This is the dominant cost for streaming analytics — evaluate whether batch Athena queries can serve the same purpose at lower cost.

### Cost Comparison: Streaming vs. Batch

| Approach | Monthly cost (10 GB/day) | Latency |
|---|---|---|
| Firehose → S3 → Athena (batch, hourly) | ~$2–5 | ~1 hour |
| Firehose → S3 + Data Analytics alerts | ~$85–100 | ~60 seconds |
| Firehose → S3 + Lambda consumer | ~$5–10 | ~60 seconds |

> **Rule of thumb:** Use Kinesis Data Analytics only when you need sub-minute windowed aggregations or stateful stream processing. For most clinical alert use cases, a Lambda triggered by Firehose S3 events is 10x cheaper.

## Section 7: CloudWatch Monitoring for Kinesis

Kinesis automatically publishes metrics to CloudWatch. Key metrics to monitor in a genomics pipeline:

**What this does:** This would call `cloudwatch.get_metric_statistics()` to retrieve the `GetRecords.IteratorAgeMilliseconds` metric for the Kinesis Data Stream over the last hour. Iterator age measures how far behind the consumer is from the most recent record — a healthy consumer should have near-zero age. If age exceeds the buffer window (e.g., >30 minutes), it indicates a consumer is too slow or has failed, and records may be at risk of expiring before being processed. This would also check `PutRecord.Success` and `ReadProvisionedThroughputExceeded` to identify throttling.

```python
import datetime

cw = boto3.client('cloudwatch', region_name=AWS_REGION)

# Retrieve iterator age metric for our stream
response = cw.get_metric_statistics(
    Namespace='AWS/Kinesis',
    MetricName='GetRecords.IteratorAgeMilliseconds',
    Dimensions=[{'Name': 'StreamName', 'Value': STREAM_NAME}],
    StartTime=datetime.datetime.utcnow() - datetime.timedelta(hours=1),
    EndTime=datetime.datetime.utcnow(),
    Period=300,         # 5-minute granularity
    Statistics=['Maximum', 'Average']
)

print(f"Iterator Age Metrics for '{STREAM_NAME}' (last 60 min):")
for dp in sorted(response['Datapoints'], key=lambda x: x['Timestamp']):
    age_sec = dp['Maximum'] / 1000
    print(f"  {dp['Timestamp'].strftime('%H:%M')}  max_age={age_sec:.1f}s  "
          f"avg_age={dp['Average']/1000:.1f}s")

# Key thresholds for a production alert:
# IteratorAge > 60000 ms → consumer is > 1 minute behind → investigate
# IteratorAge > 86400000 ms → consumer is > 24 hours behind → data loss risk
print("\nAlert threshold: IteratorAgeMilliseconds > 60,000 ms indicates consumer lag.")
```

## Section 8: Cleanup

Kinesis Data Streams charges by the **shard-hour** — even when idle. Always delete streams when the exercise is done.

**What this does:** This would call `firehose.delete_delivery_stream()` and `kinesis.delete_stream()` to remove both streaming resources. Firehose deletion is synchronous; Kinesis stream deletion is asynchronous and may take 30–60 seconds. The `enforce_consumer_deletion=True` flag on the stream delete ensures any registered enhanced fan-out consumers are also deregistered. After deletion, no further shard-hour charges accrue.

```python
# 1. Delete the Firehose delivery stream
try:
    firehose.delete_delivery_stream(DeliveryStreamName=FIREHOSE_NAME)
    print(f"Deleted Firehose delivery stream: {FIREHOSE_NAME}")
except firehose.exceptions.ResourceNotFoundException:
    print(f"Firehose '{FIREHOSE_NAME}' not found (already deleted).")

# 2. Delete the Kinesis Data Stream
try:
    kinesis.delete_stream(
        StreamName=STREAM_NAME,
        EnforceConsumerDeletion=True   # also removes enhanced fan-out consumers
    )
    print(f"Deleted Kinesis stream: {STREAM_NAME}")
    print("Note: Kinesis stream deletion takes 30-60 seconds to complete.")
except kinesis.exceptions.ResourceNotFoundException:
    print(f"Stream '{STREAM_NAME}' not found (already deleted).")

print("\nStreaming resources deleted. No further shard-hour charges.")
print("S3 objects (bronze/streaming-variants/) are retained — delete separately if needed.")
```
