## Part 1: Bronze Zone — Data Landing

We stage three TCGA file types that represent the multi-omics landscape:
- **RNA-seq HTSeq counts** — gene expression raw counts
- **Somatic mutations** — MAF format annotated variants
- **Clinical metadata** — patient demographics and outcomes

In a real deployment, you would copy files directly from `s3://tcga-2-open` using the GDC manifest. For the capstone, we generate data that exactly mirrors TCGA schemas at realistic scale.

```python
# Generate realistic TCGA BRCA multi-omics data
np.random.seed(42)

# --- Sample manifest: 30 BRCA patients, tumor + normal pairs ---
n_patients = 30

cases = []
for i in range(n_patients):
    pid = f'TCGA-{chr(65+i//10)}{chr(65+(i%10))}-{1000+i:04d}'
    age_days = np.random.randint(15000, 28000)
    stage = np.random.choice(['Stage I', 'Stage II', 'Stage III', 'Stage IV'],
                              p=[0.2, 0.45, 0.25, 0.10])
    vital = np.random.choice(['Alive', 'Dead'], p=[0.72, 0.28])
    cases.append({
        'case_id': pid,
        'project_id': 'TCGA-BRCA',
        'primary_diagnosis': np.random.choice(
            ['Infiltrating duct carcinoma, NOS', 'Lobular carcinoma, NOS',
             'Mucinous adenocarcinoma', 'Metaplastic carcinoma'],
            p=[0.75, 0.15, 0.05, 0.05]
        ),
        'tumor_stage': stage,
        'age_at_diagnosis_days': int(age_days),
        'age_at_diagnosis_years': round(age_days / 365.25, 1),
        'gender': 'female',
        'race': np.random.choice(['white', 'black or african american', 'asian', 'not reported'],
                                  p=[0.60, 0.20, 0.10, 0.10]),
        'vital_status': vital,
        'days_to_death': int(np.random.randint(100, 2000)) if vital == 'Dead' else None,
        'days_to_last_follow_up': int(np.random.randint(365, 5000)),
        'er_status': np.random.choice(['Positive', 'Negative'], p=[0.75, 0.25]),
        'her2_status': np.random.choice(['Positive', 'Negative', 'Equivocal'], p=[0.20, 0.70, 0.10]),
        'pr_status': np.random.choice(['Positive', 'Negative'], p=[0.65, 0.35]),
    })

df_clinical = pd.DataFrame(cases)
print(f"Clinical data: {len(df_clinical)} BRCA patients")
print(f"\nTumor stage distribution:")
print(df_clinical['tumor_stage'].value_counts().to_string())
```

```python
# Generate RNA-seq expression data for tumor samples
CANCER_GENES = {
    'ENSG00000012048': ('BRCA1', 81.19),
    'ENSG00000139618': ('BRCA2', 84.19),
    'ENSG00000171862': ('PTEN',  9.73),
    'ENSG00000141736': ('ERBB2', 40.26),
    'ENSG00000141510': ('TP53',  22.15),
    'ENSG00000196074': ('PIK3CA', 101.5),
    'ENSG00000000003': ('TSPAN6', 4.535),
    'ENSG00000001036': ('FUCA2',  5.621),
    'ENSG00000001084': ('GCLC',   22.807),
    'ENSG00000002330': ('BAD',    2.136),
    'ENSG00000026025': ('VIM',    8.45),
    'ENSG00000115414': ('FN1',    73.89),
    'ENSG00000166033': ('HTRA1', 47.23),
    'ENSG00000171094': ('ALK',   728.4),
    'ENSG00000105221': ('AKT2',  25.31),
}

expression_records = []
for _, case in df_clinical.iterrows():
    for gene_id, (gene_name, gene_len_kb) in CANCER_GENES.items():
        # Tumor-type-specific expression
        base_count = np.random.negative_binomial(20, 0.3)
        # Add cancer-specific signal
        if gene_name == 'ERBB2' and case['her2_status'] == 'Positive':
            base_count = int(base_count * np.random.uniform(5, 20))
        elif gene_name == 'TP53' and np.random.random() < 0.4:  # TP53 mutation
            base_count = int(base_count * np.random.uniform(0.1, 0.3))  # loss of expression
        
        expression_records.append({
            'gene_id': gene_id,
            'gene_name': gene_name,
            'gene_length_kb': gene_len_kb,
            'case_id': case['case_id'],
            'sample_type': 'Tumor',
            'raw_count': int(base_count)
        })

df_expression = pd.DataFrame(expression_records)
print(f"Expression data: {len(df_expression):,} gene-sample pairs")
print(f"Genes: {df_expression['gene_id'].nunique()}, Samples: {df_expression['case_id'].nunique()}")
```

```python
# Generate somatic mutation data (MAF-like)
DRIVER_GENES = ['TP53', 'PIK3CA', 'CDH1', 'GATA3', 'MAP3K1', 'PTEN', 'AKT1',
                 'CTCF', 'RB1', 'BRCA1', 'BRCA2', 'ERBB2', 'MYC', 'CCND1']
VARIANT_CLASSES = ['Missense_Mutation', 'Nonsense_Mutation', 'Frame_Shift_Del',
                    'Frame_Shift_Ins', 'Splice_Site', 'Silent', 'In_Frame_Del']
CHROMOSOMES = ['chr1', 'chr2', 'chr3', 'chr5', 'chr7', 'chr10', 'chr13',
                'chr16', 'chr17', 'chr20', 'chrX']

mutation_records = []
for _, case in df_clinical.iterrows():
    # Higher stage = more mutations (TMB correlation)
    stage_mult = {'Stage I': 1, 'Stage II': 2, 'Stage III': 3, 'Stage IV': 5}[case['tumor_stage']]
    n_mutations = np.random.poisson(4 * stage_mult)
    
    for _ in range(n_mutations):
        gene = np.random.choice(DRIVER_GENES + ['UNKNOWN'] * 5)  # mix driver and passenger
        chrom = np.random.choice(CHROMOSOMES)
        pos = np.random.randint(1_000_000, 200_000_000)
        t_depth = np.random.randint(30, 200)
        t_alt = np.random.randint(5, t_depth)
        
        mutation_records.append({
            'hugo_symbol': gene,
            'chromosome': chrom,
            'start_position': pos,
            'end_position': pos,
            'variant_classification': np.random.choice(VARIANT_CLASSES, p=[0.45,0.10,0.08,0.05,0.07,0.20,0.05]),
            'variant_type': np.random.choice(['SNP', 'SNP', 'SNP', 'DEL', 'INS']),
            'reference_allele': np.random.choice(['A','T','G','C']),
            'tumor_seq_allele': np.random.choice(['A','T','G','C']),
            'tumor_sample_barcode': f"{case['case_id']}-01A",
            'case_id': case['case_id'],
            't_depth': int(t_depth),
            't_ref_count': int(t_depth - t_alt),
            't_alt_count': int(t_alt),
            'vaf': round(t_alt / t_depth, 3)
        })

df_mutations = pd.DataFrame(mutation_records)
print(f"Somatic mutations: {len(df_mutations):,} total across {df_clinical['case_id'].nunique()} patients")
print(f"\nTop mutated genes:")
print(df_mutations[df_mutations['variant_classification'] != 'Silent']
      ['hugo_symbol'].value_counts().head(8).to_string())
```

```python
# Write all three data types to S3 bronze zone

# Clinical
s3.put_object(
    Bucket=BUCKET,
    Key='bronze/clinical/cancer_type=BRCA/clinical.tsv',
    Body=df_clinical.to_csv(sep='\t', index=False).encode()
)

# Expression (one file per sample — as TCGA delivers it)
for case_id, group in df_expression.groupby('case_id'):
    htseq = group[['gene_id', 'raw_count']].copy()
    s3.put_object(
        Bucket=BUCKET,
        Key=f'bronze/expression/cancer_type=BRCA/case_id={case_id}/htseq_counts.txt',
        Body=htseq.to_csv(sep='\t', index=False, header=False).encode()
    )

# Mutations (one MAF per cancer type)
maf_header = '#version 2.4\n'
s3.put_object(
    Bucket=BUCKET,
    Key='bronze/mutations/cancer_type=BRCA/somatic_mutations.maf',
    Body=(maf_header + df_mutations.to_csv(sep='\t', index=False)).encode()
)

print("Bronze zone written:")
print(f"  Clinical: 1 TSV file ({len(df_clinical)} patients)")
print(f"  Expression: {df_expression['case_id'].nunique()} HTSeq files")
print(f"  Mutations: 1 MAF file ({len(df_mutations)} mutations)")
```

## Part 2: Silver Zone — Cleaning and Standardizing

Apply the transformations we practiced in Exercise 4:
- RNA-seq: normalize raw counts to TPM, melt wide to long format
- Mutations: clean MAF, write partitioned by chromosome
- Clinical: standardize column names, resolve missing values

```python
# Silver: RNA-seq — TPM normalization and long format
gene_lengths = dict(zip(
    [g for g, _ in CANCER_GENES.values()],
    [l for _, l in CANCER_GENES.values()]
))
# Fix: key by gene_id, not gene_name
gene_lengths_by_id = {
    gene_id: gene_len
    for gene_id, (gene_name, gene_len) in CANCER_GENES.items()
}

def compute_tpm(counts, lengths):
    """Compute TPM from raw counts and gene lengths in kb."""
    rate = counts / lengths
    return (rate / rate.sum()) * 1e6

# Compute TPM per sample
tpm_records = []
for case_id, group in df_expression.groupby('case_id'):
    g = group.set_index('gene_id')
    lengths = g['gene_length_kb']
    counts = g['raw_count'].astype(float)
    tpm = compute_tpm(counts, lengths)
    
    for gene_id, tpm_val in tpm.items():
        gene_row = g.loc[gene_id]
        tpm_records.append({
            'gene_id': gene_id,
            'gene_name': gene_row['gene_name'],
            'case_id': case_id,
            'sample_type': gene_row['sample_type'],
            'raw_count': int(gene_row['raw_count']),
            'tpm': round(float(tpm_val), 4),
            'cancer_type': 'BRCA'
        })

df_silver_expression = pd.DataFrame(tpm_records)

wr.s3.to_parquet(
    df=df_silver_expression,
    path=f"s3://{BUCKET}/silver/expression/",
    dataset=True,
    database='binfx410_silver',
    table='tcga_expression_tpm',
    partition_cols=['cancer_type', 'sample_type'],
    boto3_session=session,
    mode='overwrite',
    compression='snappy'
)
print(f"Silver expression: {len(df_silver_expression):,} rows — partitioned by cancer_type/sample_type")
```

```python
# Silver: Somatic mutations — clean and partition by chromosome
df_silver_mutations = df_mutations.copy()

# Clean: remove chr prefix for uniformity, filter Silent mutations for most analyses
df_silver_mutations['chromosome_clean'] = df_silver_mutations['chromosome'].str.replace('chr', '', regex=False)
df_silver_mutations['is_driver_gene'] = df_silver_mutations['hugo_symbol'].isin(DRIVER_GENES)
df_silver_mutations['is_functional'] = df_silver_mutations['variant_classification'] != 'Silent'
df_silver_mutations['cancer_type'] = 'BRCA'

wr.s3.to_parquet(
    df=df_silver_mutations,
    path=f"s3://{BUCKET}/silver/mutations/",
    dataset=True,
    database='binfx410_silver',
    table='tcga_somatic_mutations',
    partition_cols=['cancer_type', 'chromosome'],
    boto3_session=session,
    mode='overwrite',
    compression='snappy'
)
print(f"Silver mutations: {len(df_silver_mutations):,} rows — partitioned by cancer_type/chromosome")
```

```python
# Silver: Clinical — standardize and clean
df_silver_clinical = df_clinical.copy()

# Standardize: replace None with proper NaN, standardize stage to Roman numerals
df_silver_clinical['stage_roman'] = df_silver_clinical['tumor_stage'].str.extract(r'Stage (\S+)')[0]
df_silver_clinical['is_her2_positive'] = df_silver_clinical['her2_status'] == 'Positive'
df_silver_clinical['is_hr_positive'] = (
    (df_silver_clinical['er_status'] == 'Positive') |
    (df_silver_clinical['pr_status'] == 'Positive')
)
df_silver_clinical['cancer_type'] = 'BRCA'

wr.s3.to_parquet(
    df=df_silver_clinical,
    path=f"s3://{BUCKET}/silver/clinical/",
    dataset=True,
    database='binfx410_silver',
    table='tcga_clinical',
    partition_cols=['cancer_type'],
    boto3_session=session,
    mode='overwrite'
)
print(f"Silver clinical: {len(df_silver_clinical)} patients — partitioned by cancer_type")
```

## Part 3: Gold Zone — Iceberg Tables with Analytics Aggregations

```python
# Compute Tumor Mutation Burden (TMB)
# TMB = number of nonsynonymous mutations / megabases of genome sequenced
# WES captures ~35 Mb of coding sequence

EXOME_SIZE_MB = 35.0  # typical WES target size

tmb_df = (df_silver_mutations[df_silver_mutations['is_functional']]
          .groupby('case_id')
          .size()
          .reset_index(name='functional_mutation_count'))

tmb_df['tmb_per_mb'] = (tmb_df['functional_mutation_count'] / EXOME_SIZE_MB).round(2)
tmb_df['tmb_category'] = pd.cut(
    tmb_df['tmb_per_mb'],
    bins=[0, 5, 10, 20, np.inf],
    labels=['Low (<5)', 'Intermediate (5-10)', 'High (10-20)', 'Very High (>20)']
).astype(str)

# Join to clinical for survival data
tmb_with_clinical = tmb_df.merge(
    df_silver_clinical[['case_id', 'tumor_stage', 'stage_roman', 'vital_status',
                         'days_to_death', 'days_to_last_follow_up', 'is_her2_positive',
                         'is_hr_positive', 'er_status', 'cancer_type']],
    on='case_id'
)

print("Tumor Mutation Burden per patient:")
print(tmb_df.sort_values('tmb_per_mb', ascending=False).head(10).to_string(index=False))
print(f"\nTMB category distribution:")
print(tmb_df['tmb_category'].value_counts().to_string())
```

```python
# Write Gold TMB table — use Iceberg for ACID (TCGA releases corrections)
wr.s3.to_parquet(
    df=tmb_with_clinical,
    path=f"s3://{BUCKET}/gold/tmb_per_patient/",
    dataset=True,
    database='binfx410_gold',
    table='tmb_per_patient',
    boto3_session=session,
    mode='overwrite'
)

# For Iceberg version (requires Athena v3)
def run_sql(sql, db='binfx410_gold'):
    qid = athena.start_query_execution(
        QueryString=sql,
        QueryExecutionContext={'Database': db},
        ResultConfiguration={'OutputLocation': ATHENA_RESULTS},
        WorkGroup='primary'
    )['QueryExecutionId']
    for _ in range(60):
        time.sleep(2)
        state = athena.get_query_execution(QueryExecutionId=qid)['QueryExecution']['Status']['State']
        if state in ('SUCCEEDED', 'FAILED', 'CANCELLED'):
            break
    return qid

print("Gold TMB table written.")
```

```python
# Compute expression signatures: top 5 most variable genes per cancer type
gene_variability = (
    df_silver_expression
    .groupby(['gene_id', 'gene_name', 'cancer_type'])['tpm']
    .agg(mean_tpm='mean', std_tpm='std', min_tpm='min', max_tpm='max')
    .reset_index()
)
gene_variability['cv'] = (gene_variability['std_tpm'] / gene_variability['mean_tpm']).fillna(0)
gene_variability['is_hvg'] = gene_variability.groupby('cancer_type')['cv'].transform(
    lambda x: x >= x.quantile(0.8)
)

wr.s3.to_parquet(
    df=gene_variability,
    path=f"s3://{BUCKET}/gold/expression_signatures/",
    dataset=True,
    database='binfx410_gold',
    table='expression_signatures',
    partition_cols=['cancer_type'],
    boto3_session=session,
    mode='overwrite'
)
print("Gold expression signatures written.")
print(f"HVGs identified: {gene_variability['is_hvg'].sum()} out of {len(gene_variability)} gene-cancer combinations")
```

## Part 4: Star Schema Analytical Layer

```python
# Build the star schema using CTAS

try:
    glue.create_database(DatabaseInput={'Name': 'binfx410_mart'})
except: pass

# fact_patient_molecular_profile
fact_df = tmb_with_clinical.merge(
    gene_variability[gene_variability['is_hvg']]
    .groupby('cancer_type')['gene_name']
    .apply(lambda x: ','.join(x[:5]))
    .reset_index(name='top_hvg_genes'),
    on='cancer_type',
    how='left'
)

# dim_patient
dim_patient_df = df_silver_clinical[[
    'case_id', 'tumor_stage', 'stage_roman', 'vital_status',
    'days_to_last_follow_up', 'days_to_death', 'race',
    'er_status', 'her2_status', 'pr_status', 'is_her2_positive', 'is_hr_positive'
]].rename(columns={'case_id': 'patient_key'})

# dim_gene
dim_gene_df = gene_variability[['gene_id', 'gene_name', 'cancer_type']].drop_duplicates()
dim_gene_df.columns = ['gene_key', 'gene_name', 'cancer_type']

for table_name, df in [
    ('fact_patient_molecular', fact_df),
    ('dim_patient', dim_patient_df),
    ('dim_gene', dim_gene_df)
]:
    wr.s3.to_parquet(
        df=df,
        path=f"s3://{BUCKET}/gold/mart/{table_name}/",
        dataset=True,
        database='binfx410_mart',
        table=table_name,
        boto3_session=session,
        mode='overwrite'
    )
    print(f"Star schema table written: {table_name} ({len(df)} rows)")
```

## Part 5: Analytical Queries

```python
# Query 1: TMB vs tumor stage relationship
q1 = wr.athena.read_sql_query(
    sql="""
    SELECT
        tumor_stage,
        COUNT(*) AS patient_count,
        ROUND(AVG(tmb_per_mb), 2) AS avg_tmb,
        ROUND(MIN(tmb_per_mb), 2) AS min_tmb,
        ROUND(MAX(tmb_per_mb), 2) AS max_tmb
    FROM binfx410_gold.tmb_per_patient
    GROUP BY tumor_stage
    ORDER BY tumor_stage
    """,
    database='binfx410_gold',
    s3_output=ATHENA_RESULTS,
    boto3_session=session
)
print("Q1: TMB by tumor stage")
q1
```

```python
# Query 2: HER2+ patients — do they have higher ERBB2 expression?
q2_df = df_silver_expression[df_silver_expression['gene_name'] == 'ERBB2'].merge(
    df_silver_clinical[['case_id', 'her2_status']],
    on='case_id'
)

erbb2_by_her2 = q2_df.groupby('her2_status')['tpm'].agg(
    mean_tpm='mean', median_tpm='median', std_tpm='std', n='count'
).round(2)

print("Q2: ERBB2 (HER2 gene) expression by HER2 protein status:")
print(erbb2_by_her2.to_string())
print()
print("Expected: HER2+ tumors should have much higher ERBB2 TPM")
```

```python
# Query 3: Top mutated genes in High TMB tumors vs Low TMB
high_tmb_cases = tmb_df[tmb_df['tmb_per_mb'] > 10]['case_id'].tolist()
low_tmb_cases = tmb_df[tmb_df['tmb_per_mb'] <= 5]['case_id'].tolist()

def top_genes(case_ids, label):
    subset = df_silver_mutations[
        df_silver_mutations['case_id'].isin(case_ids) &
        (df_silver_mutations['variant_classification'] != 'Silent') &
        df_silver_mutations['is_driver_gene']
    ]
    freq = subset['hugo_symbol'].value_counts().head(8).reset_index()
    freq.columns = ['gene', 'mutation_count']
    freq['tmb_group'] = label
    freq['freq_pct'] = (freq['mutation_count'] / max(len(case_ids), 1) * 100).round(1)
    return freq

high_tmb_genes = top_genes(high_tmb_cases, 'High TMB')
low_tmb_genes = top_genes(low_tmb_cases, 'Low TMB')

print("Q3: Top mutated driver genes by TMB group")
print("\nHigh TMB tumors (TMB > 10):")
print(high_tmb_genes.to_string(index=False))
print("\nLow TMB tumors (TMB <= 5):")
print(low_tmb_genes.to_string(index=False))
```

```python
# Visualization: TMB vs HER2 status + survival
fig, axes = plt.subplots(1, 3, figsize=(16, 5))
fig.suptitle('TCGA BRCA — Multi-Omics Lakehouse Analysis\n(Gold Layer Queries)',
             fontsize=12, fontweight='bold')

# Plot 1: TMB by stage
ax1 = axes[0]
stage_order = ['Stage I', 'Stage II', 'Stage III', 'Stage IV']
stage_tmb = tmb_with_clinical.groupby('tumor_stage')['tmb_per_mb'].mean().reindex(stage_order)
colors = ['#4CAF50', '#FF9800', '#F44336', '#9C27B0']
ax1.bar(stage_order, stage_tmb, color=colors)
ax1.set_title('Mean TMB by Stage', fontweight='bold')
ax1.set_ylabel('TMB (mutations/Mb)')
ax1.tick_params(axis='x', rotation=20)

# Plot 2: ERBB2 expression by HER2 status
ax2 = axes[1]
her2_groups = [q2_df[q2_df['her2_status'] == s]['tpm'].values
                for s in ['Negative', 'Equivocal', 'Positive']]
bp = ax2.boxplot(her2_groups, labels=['HER2-', 'Equivocal', 'HER2+'],
                  patch_artist=True)
for patch, color in zip(bp['boxes'], ['#4CAF50', '#FF9800', '#F44336']):
    patch.set_facecolor(color)
ax2.set_title('ERBB2 Expression by HER2 Status', fontweight='bold')
ax2.set_ylabel('TPM')

# Plot 3: TMB distribution overall
ax3 = axes[2]
ax3.hist(tmb_df['tmb_per_mb'], bins=15, color='#2196F3', edgecolor='white')
ax3.axvline(x=10, color='red', linestyle='--', label='High TMB threshold')
ax3.set_title('TMB Distribution — BRCA', fontweight='bold')
ax3.set_xlabel('Mutations per Megabase')
ax3.set_ylabel('Patient Count')
ax3.legend()

plt.tight_layout()
plt.savefig('capstone_analysis.png', dpi=150, bbox_inches='tight')
plt.show()
```

## Part 6: Apply Lake Formation Governance

```python
# Apply column exclusion on clinical table
# The 'days_to_death' and 'age_at_diagnosis_days' are quasi-PHI
# (combined with other fields, these can identify individuals)

QUASI_PHI_COLS = ['days_to_death', 'age_at_diagnosis_days', 'days_to_last_follow_up']

try:
    ACCOUNT_ID = sts.get_caller_identity()['Account']
    
    # Create a public-researcher role
    trust_policy = json.dumps({
        "Version": "2012-10-17",
        "Statement": [{
            "Effect": "Allow",
            "Principal": {"AWS": sts.get_caller_identity()['Arn']},
            "Action": "sts:AssumeRole"
        }]
    })
    
    try:
        iam.create_role(
            RoleName='binfx410-public-researcher',
            AssumeRolePolicyDocument=trust_policy
        )
    except iam.exceptions.EntityAlreadyExistsException:
        pass
    
    PUBLIC_ARN = f'arn:aws:iam::{ACCOUNT_ID}:role/binfx410-public-researcher'
    
    # Grant with PHI exclusion
    lf.grant_permissions(
        Principal={'DataLakePrincipalIdentifier': PUBLIC_ARN},
        Resource={
            'TableWithColumns': {
                'CatalogId': ACCOUNT_ID,
                'DatabaseName': 'binfx410_silver',
                'Name': 'tcga_clinical',
                'ColumnWildcard': {
                    'ExcludedColumnNames': QUASI_PHI_COLS
                }
            }
        },
        Permissions=['SELECT']
    )
    print(f"Lake Formation: public-researcher role excludes quasi-PHI columns:")
    for c in QUASI_PHI_COLS:
        print(f"  - {c}")
except Exception as e:
    print(f"Note (LF may need admin setup): {e}")
```
