## Background: Sequencing Instrument Data Streams

### What Modern Sequencers Emit

Illumina sequencers continuously emit run QC metrics during sequencing:

```json
{
  "run_id": "RUN_20260315_001",
  "instrument": "NovaSeq 6000",
  "flow_cell_id": "FC-240315001",
  "sample_id": "SMP-A42",
  "bases_q30_pct": 94.2,
  "cluster_density": 182000,
  "yield_gb": 312.5,
  "error_rate_pct": 0.18,
  "completed_at": "2026-03-15T14:22:00Z"
}
```

A busy genomics core facility runs 5–20 flow cells per week. For real-time monitoring, labs need:

1. **Hot path (immediate):** Alert when Q30 < 80% (sequencing run failed)
2. **Cold path (historical):** Trend analysis — yield trends by instrument, seasonal patterns

### Why Not Just Write to S3 Directly?

If each run completion event writes one JSON file to S3:
- 20 runs/week × 52 weeks = 1,040 tiny JSON files/year
- Athena must open/close each file separately
- Each file is ~200 bytes — massive overhead per byte of useful data

**Kinesis Firehose** solves this: it **micro-batches** incoming events, waits 60 seconds or until 5 MB accumulates, then writes one efficient Parquet file — and automatically converts JSON to Parquet using the Glue schema.

````python
import boto3
import awswrangler as wr
import pandas as pd
import json
import base64
import time
import random
import numpy as np
from datetime import datetime, timezone, timedelta
import matplotlib.pyplot as plt

STUDENT_ID = "your-student-id"  # CHANGE THIS
AWS_REGION = "us-east-1"
BUCKET = f"binfx410-datalake-{STUDENT_ID}"
ATHENA_RESULTS = f"s3://binfx410-athena-results-{STUDENT_ID}/"
FIREHOSE_STREAM = f"binfx410-sequencer-feed-{STUDENT_ID}"

s3 = boto3.client('s3', region_name=AWS_REGION)
glue = boto3.client('glue', region_name=AWS_REGION)
firehose = boto3.client('firehose', region_name=AWS_REGION)
iam = boto3.client('iam')
sts = boto3.client('sts')
session = boto3.Session(region_name=AWS_REGION)

ACCOUNT_ID = sts.get_caller_identity()['Account']
print(f"Account: {ACCOUNT_ID}")
print(f"Firehose stream: {FIREHOSE_STREAM}")
````

## Section 2: Defining the Event Schema

````python
# Define the sequencer run completion event schema
# This mirrors the structure of Illumina BaseSpace run completion events

INSTRUMENTS = [
    'NovaSeq 6000', 'NovaSeq X Plus', 'NextSeq 2000',
    'MiSeq', 'HiSeq 2500'
]
LABS = ['Genomics Core A', 'Genomics Core B', 'Clinical Lab', 'Research Lab 3']
SEQUENCING_TYPES = ['WGS-30x', 'WGS-10x', 'WES', 'RNA-seq', 'ChIP-seq', 'ATAC-seq']

random.seed(int(time.time()))  # use real randomness for simulation

def generate_run_event():
    """Generate a realistic sequencing run completion event."""
    instrument = random.choice(INSTRUMENTS)
    
    # Instrument-specific parameters
    if 'NovaSeq' in instrument:
        max_yield = 3000
        typical_clusters = random.randint(160000, 220000)
    elif 'NextSeq' in instrument:
        max_yield = 300
        typical_clusters = random.randint(100000, 160000)
    else:  # MiSeq/HiSeq
        max_yield = 15
        typical_clusters = random.randint(800, 2000)
    
    # Occasionally simulate a failed run (Q30 < 80%)
    is_failed_run = random.random() < 0.08  # 8% failure rate
    q30 = round(random.uniform(60, 78), 2) if is_failed_run else round(random.uniform(82, 97), 2)
    yield_gb = round(random.uniform(max_yield * 0.7, max_yield), 1)
    
    now = datetime.now(timezone.utc)
    
    return {
        'run_id': f"RUN_{now.strftime('%Y%m%d_%H%M%S')}_{random.randint(100,999)}",
        'instrument_model': instrument,
        'instrument_serial': f"SN-{random.randint(10000, 99999)}",
        'flow_cell_id': f"FC-{random.randint(100000, 999999)}",
        'sample_id': f"SMP-{random.randint(1000, 9999)}",
        'sequencing_type': random.choice(SEQUENCING_TYPES),
        'lab_id': random.choice(LABS),
        'bases_q30_pct': q30,
        'error_rate_pct': round(random.uniform(0.1, 0.5 if not is_failed_run else 2.5), 3),
        'cluster_density': typical_clusters,
        'yield_gb': yield_gb,
        'read_length': random.choice([75, 100, 150, 250]),
        'num_reads': round(yield_gb * 1e9 / random.choice([75, 100, 150, 250]) / 2, 0),
        'run_duration_hours': round(random.uniform(8, 60), 1),
        'is_paired_end': random.choice([True, True, False]),
        'run_quality': 'FAILED' if is_failed_run else 'PASSED',
        'completed_at': now.strftime('%Y-%m-%dT%H:%M:%SZ'),
        'dt': now.strftime('%Y-%m-%d')  # partition key for Firehose
    }

# Show a sample event
sample_event = generate_run_event()
print("Sample sequencer event:")
print(json.dumps(sample_event, indent=2))
````

## Section 3: Creating the Glue Schema for Firehose Parquet Conversion

````python
# Firehose needs a Glue table definition to know how to convert JSON → Parquet
# The schema must exactly match the event JSON structure

try:
    glue.create_database(DatabaseInput={'Name': 'binfx410_streaming'})
    print("Created database: binfx410_streaming")
except glue.exceptions.AlreadyExistsException:
    print("Database binfx410_streaming already exists.")

# Create Glue table with the event schema
STREAMING_COLUMNS = [
    {'Name': 'run_id',              'Type': 'string'},
    {'Name': 'instrument_model',    'Type': 'string'},
    {'Name': 'instrument_serial',   'Type': 'string'},
    {'Name': 'flow_cell_id',        'Type': 'string'},
    {'Name': 'sample_id',           'Type': 'string'},
    {'Name': 'sequencing_type',     'Type': 'string'},
    {'Name': 'lab_id',              'Type': 'string'},
    {'Name': 'bases_q30_pct',       'Type': 'double'},
    {'Name': 'error_rate_pct',      'Type': 'double'},
    {'Name': 'cluster_density',     'Type': 'bigint'},
    {'Name': 'yield_gb',            'Type': 'double'},
    {'Name': 'read_length',         'Type': 'int'},
    {'Name': 'num_reads',           'Type': 'double'},
    {'Name': 'run_duration_hours',  'Type': 'double'},
    {'Name': 'is_paired_end',       'Type': 'boolean'},
    {'Name': 'run_quality',         'Type': 'string'},
    {'Name': 'completed_at',        'Type': 'string'},
]

try:
    glue.delete_table(DatabaseName='binfx410_streaming', Name='sequencer_runs')
except: pass

glue.create_table(
    DatabaseName='binfx410_streaming',
    TableInput={
        'Name': 'sequencer_runs',
        'StorageDescriptor': {
            'Columns': STREAMING_COLUMNS,
            'Location': f's3://{BUCKET}/streaming/sequencer_runs/',
            'InputFormat': 'org.apache.hadoop.mapred.TextInputFormat',
            'OutputFormat': 'org.apache.hadoop.hive.ql.io.HiveIgnoreKeyTextOutputFormat',
            'SerdeInfo': {
                'SerializationLibrary': 'org.openx.data.jsonserde.JsonSerDe',
                'Parameters': {'serialization.format': '1'}
            }
        },
        'PartitionKeys': [{'Name': 'dt', 'Type': 'string'}]
    }
)
print(f"Glue schema created: binfx410_streaming.sequencer_runs")
print(f"  {len(STREAMING_COLUMNS)} columns defined")
print(f"  Partition key: dt (date string)")
````

## Section 4: Creating the Kinesis Firehose Delivery Stream

````python
# Create IAM role for Firehose
firehose_trust = json.dumps({
    "Version": "2012-10-17",
    "Statement": [{
        "Effect": "Allow",
        "Principal": {"Service": "firehose.amazonaws.com"},
        "Action": "sts:AssumeRole"
    }]
})

firehose_policy = json.dumps({
    "Version": "2012-10-17",
    "Statement": [
        {
            "Effect": "Allow",
            "Action": ["s3:PutObject", "s3:GetObject", "s3:ListBucket", "s3:AbortMultipartUpload"],
            "Resource": [f"arn:aws:s3:::{BUCKET}", f"arn:aws:s3:::{BUCKET}/*"]
        },
        {
            "Effect": "Allow",
            "Action": ["glue:GetTable", "glue:GetTableVersions", "glue:GetTableVersion"],
            "Resource": ["*"]
        },
        {
            "Effect": "Allow",
            "Action": ["logs:PutLogEvents", "logs:CreateLogGroup", "logs:CreateLogStream"],
            "Resource": ["arn:aws:logs:*:*:*"]
        }
    ]
})

FIREHOSE_ROLE = f"binfx410-firehose-role-{STUDENT_ID}"

try:
    iam.create_role(
        RoleName=FIREHOSE_ROLE,
        AssumeRolePolicyDocument=firehose_trust,
        Description='Firehose role for BINFX410 streaming exercise'
    )
    print(f"Created role: {FIREHOSE_ROLE}")
except iam.exceptions.EntityAlreadyExistsException:
    print(f"Role exists: {FIREHOSE_ROLE}")

iam.put_role_policy(
    RoleName=FIREHOSE_ROLE,
    PolicyName='FirehoseS3GlueAccess',
    PolicyDocument=firehose_policy
)

FIREHOSE_ROLE_ARN = f'arn:aws:iam::{ACCOUNT_ID}:role/{FIREHOSE_ROLE}'
print(f"Role ARN: {FIREHOSE_ROLE_ARN}")
print("Waiting 15 seconds for IAM propagation...")
time.sleep(15)
````

````python
# Create the Firehose delivery stream
# Configured to: buffer 60s or 5MB, convert JSON→Parquet using Glue schema

try:
    firehose.delete_delivery_stream(DeliveryStreamName=FIREHOSE_STREAM, AllowForceDelete=True)
    time.sleep(5)
except: pass

create_response = firehose.create_delivery_stream(
    DeliveryStreamName=FIREHOSE_STREAM,
    DeliveryStreamType='DirectPut',
    ExtendedS3DestinationConfiguration={
        'RoleARN': FIREHOSE_ROLE_ARN,
        'BucketARN': f'arn:aws:s3:::{BUCKET}',
        'Prefix': 'streaming/sequencer_runs/dt=!{partitionKeyFromSchema:dt}/',
        'ErrorOutputPrefix': 'streaming/errors/',
        'BufferingHints': {
            'SizeInMBs': 5,        # flush when 5 MB accumulated
            'IntervalInSeconds': 60  # or every 60 seconds, whichever comes first
        },
        'CompressionFormat': 'UNCOMPRESSED',  # Parquet handles its own compression
        'DataFormatConversionConfiguration': {
            'SchemaConfiguration': {
                'RoleARN': FIREHOSE_ROLE_ARN,
                'DatabaseName': 'binfx410_streaming',
                'TableName': 'sequencer_runs',
                'Region': AWS_REGION,
                'VersionId': 'LATEST'
            },
            'InputFormatConfiguration': {
                'Deserializer': {'OpenXJsonSerDe': {}}
            },
            'OutputFormatConfiguration': {
                'Serializer': {
                    'ParquetSerDe': {
                        'Compression': 'SNAPPY'
                    }
                }
            },
            'Enabled': True
        },
        'DynamicPartitioningConfiguration': {
            'Enabled': True,
            'RetryOptions': {'DurationInSeconds': 300}
        }
    }
)

print(f"Firehose stream created: {FIREHOSE_STREAM}")
print(f"  Buffer: 60 seconds OR 5 MB (whichever first)")
print(f"  Format conversion: JSON → Parquet (Snappy)")
print(f"  Destination: s3://{BUCKET}/streaming/sequencer_runs/dt=YYYY-MM-DD/")

# Wait for stream to become ACTIVE
for i in range(20):
    status = firehose.describe_delivery_stream(DeliveryStreamName=FIREHOSE_STREAM)
    state = status['DeliveryStreamDescription']['DeliveryStreamStatus']
    if state == 'ACTIVE':
        print(f"  Stream is ACTIVE and ready.")
        break
    print(f"  [{i*5}s] Status: {state}...")
    time.sleep(5)
````

## Section 5: Simulating Sequencer Events

````python
# Lambda handler code (shown for educational purposes)
# In production, this would be deployed as a Lambda triggered by the sequencer

lambda_handler_code = '''
import json
import boto3
import random
from datetime import datetime, timezone

firehose = boto3.client('firehose', region_name='us-east-1')
sns = boto3.client('sns', region_name='us-east-1')

FIREHOSE_STREAM = 'binfx410-sequencer-feed'
ALERT_TOPIC_ARN = 'arn:aws:sns:us-east-1:ACCOUNT_ID:sequencer-alerts'
Q30_THRESHOLD = 80.0  # Alert when Q30% drops below this

def lambda_handler(event, context):
    """
    Called when a sequencer run completes.
    1. Sends event to Firehose (cold path — for historical analysis)
    2. Checks Q30% — if below threshold, sends SNS alert (hot path)
    """
    run_data = event.get(\'run_data\', {})
    
    # HOT PATH: Real-time alert for failed runs
    if run_data.get(\'bases_q30_pct\', 100) < Q30_THRESHOLD:
        sns.publish(
            TopicArn=ALERT_TOPIC_ARN,
            Subject=f"SEQUENCER ALERT: Run {run_data[\'run_id\']} FAILED",
            Message=f"Q30% = {run_data[\'bases_q30_pct\']}% on {run_data[\'instrument_model\']}"
        )
    
    # COLD PATH: Send to Firehose for batching → Parquet → Athena
    firehose.put_record(
        DeliveryStreamName=FIREHOSE_STREAM,
        Record={\'Data\': json.dumps(run_data) + \'\\n\'}  # newline-delimited JSON
    )
    
    return {\'status\': \'ok\', \'run_id\': run_data.get(\'run_id\')}
'''

print("Lambda handler (hot path + cold path):")
print(lambda_handler_code)
````

````python
# Send 3 batches of events to Firehose
# Each batch simulates one hour of sequencer activity in a busy core facility

all_sent_events = []
failed_runs = []

for batch_num in range(3):
    print(f"\nBatch {batch_num + 1}/3 — sending 15 events...")
    
    # Generate 15 events per batch
    records = []
    batch_events = []
    for _ in range(15):
        event = generate_run_event()
        batch_events.append(event)
        all_sent_events.append(event)
        
        if event['run_quality'] == 'FAILED':
            failed_runs.append(event)
        
        # Firehose record: newline-delimited JSON
        record_data = json.dumps(event) + '\n'
        records.append({'Data': record_data.encode()})
    
    # Send batch to Firehose
    response = firehose.put_record_batch(
        DeliveryStreamName=FIREHOSE_STREAM,
        Records=records
    )
    
    failed_count = response.get('FailedPutCount', 0)
    print(f"  Sent: {len(records)}, Failed: {failed_count}")
    
    # HOT PATH simulation: alert on Q30 failures immediately
    for event in batch_events:
        if event['run_quality'] == 'FAILED':
            print(f"  *** ALERT: {event['run_id']} FAILED — Q30={event['bases_q30_pct']}% ***")
    
    if batch_num < 2:
        print(f"  Waiting 30 seconds before next batch...")
        time.sleep(30)

df_sent = pd.DataFrame(all_sent_events)
print(f"\nTotal events sent: {len(df_sent)}")
print(f"Failed runs detected: {len(failed_runs)}")
print()
print("NOTE: Firehose buffers for 60 seconds. Wait before querying.")
````

## Section 6: Waiting for Firehose to Flush and Querying Results

````python
# Firehose buffers for up to 60 seconds — wait for flush
print("Waiting 75 seconds for Firehose buffer to flush...")
for i in range(75, 0, -5):
    print(f"  {i}s remaining...", end='\r')
    time.sleep(5)
print("\nFlush complete. Checking S3 for Parquet files...")

# List files in the streaming output
paginator = s3.get_paginator('list_objects_v2')
streaming_files = []
for page in paginator.paginate(Bucket=BUCKET, Prefix='streaming/sequencer_runs/'):
    for obj in page.get('Contents', []):
        streaming_files.append({
            'key': obj['Key'],
            'size_kb': round(obj['Size'] / 1024, 1),
            'last_modified': obj['LastModified'].strftime('%Y-%m-%d %H:%M:%S')
        })

if streaming_files:
    print(f"Found {len(streaming_files)} file(s) written by Firehose:")
    for f in streaming_files:
        print(f"  {f['key'].split('/')[-1]}  ({f['size_kb']} KB)  {f['last_modified']}")
else:
    print("No files yet — Firehose may still be buffering. Try again in 30 seconds.")
````

````python
# Add partition to Athena so it can find the data
today = datetime.now().strftime('%Y-%m-%d')

try:
    wr.athena.start_query_execution(
        sql=f"ALTER TABLE binfx410_streaming.sequencer_runs ADD IF NOT EXISTS PARTITION (dt='{today}')",
        database='binfx410_streaming',
        s3_output=ATHENA_RESULTS,
        boto3_session=session,
        wait=True
    )
    print(f"Partition dt={today} added.")
except Exception as e:
    print(f"Note: {e}")
    print("If Firehose wrote data, try MSCK REPAIR TABLE instead.")
````

````python
# COLD PATH QUERY: Historical analysis — yield trend by instrument
historical_query = """
SELECT
    instrument_model,
    COUNT(*) AS run_count,
    ROUND(AVG(bases_q30_pct), 2) AS avg_q30,
    ROUND(AVG(yield_gb), 1) AS avg_yield_gb,
    SUM(CASE WHEN run_quality = 'FAILED' THEN 1 ELSE 0 END) AS failed_runs
FROM binfx410_streaming.sequencer_runs
WHERE dt = '{today}'
GROUP BY instrument_model
ORDER BY run_count DESC
""".format(today=today)

df_historical = wr.athena.read_sql_query(
    sql=historical_query,
    database='binfx410_streaming',
    s3_output=ATHENA_RESULTS,
    boto3_session=session
)
print("COLD PATH — Instrument performance summary (from Parquet in S3):")
df_historical
````

````python
# HOT PATH QUERY: Alert-worthy runs — Q30 < 80%
alert_query = """
SELECT
    run_id,
    instrument_model,
    lab_id,
    bases_q30_pct,
    error_rate_pct,
    yield_gb,
    completed_at
FROM binfx410_streaming.sequencer_runs
WHERE dt = '{today}'
  AND bases_q30_pct < 80
ORDER BY bases_q30_pct ASC
""".format(today=today)

df_alerts = wr.athena.read_sql_query(
    sql=alert_query,
    database='binfx410_streaming',
    s3_output=ATHENA_RESULTS,
    boto3_session=session
)
print(f"ALERT QUERY — Failed runs (Q30 < 80%): {len(df_alerts)} runs")
if len(df_alerts) > 0:
    df_alerts
````

````python
# Visualize the sent events vs queried events
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle('Sequencer Run Telemetry — Hot/Cold Path Analysis', fontweight='bold')

# Plot 1: Q30 distribution by instrument
ax1 = axes[0]
for instrument in df_sent['instrument_model'].unique():
    subset = df_sent[df_sent['instrument_model'] == instrument]['bases_q30_pct']
    ax1.hist(subset, alpha=0.6, label=instrument[:12], bins=10)
ax1.axvline(x=80, color='red', linestyle='--', linewidth=2, label='Alert threshold (80%)')
ax1.set_xlabel('Q30%')
ax1.set_ylabel('Run Count')
ax1.set_title('Q30% Distribution by Instrument')
ax1.legend(fontsize=7)

# Plot 2: Yield by sequencing type
ax2 = axes[1]
yield_by_type = df_sent.groupby('sequencing_type')['yield_gb'].mean().sort_values(ascending=True)
yield_by_type.plot(kind='barh', ax=ax2, color='#2196F3')
ax2.set_xlabel('Avg Yield (GB)')
ax2.set_title('Average Yield by Sequencing Type')

plt.tight_layout()
plt.savefig('sequencer_telemetry.png', dpi=150, bbox_inches='tight')
plt.show()
````

## Section 7: Comparing Direct-to-S3 vs Firehose

````python
# Simulate what would happen if we wrote each event directly as JSON to S3
# (the naive approach that causes the small-files problem)

for i, event in enumerate(all_sent_events):
    key = f"streaming/direct_json/dt={today}/event_{i:04d}_{event['run_id']}.json"
    s3.put_object(
        Bucket=BUCKET,
        Key=key,
        Body=json.dumps(event).encode()
    )

# Compare file counts and sizes
def get_prefix_stats(prefix):
    paginator = s3.get_paginator('list_objects_v2')
    files = []
    for page in paginator.paginate(Bucket=BUCKET, Prefix=prefix):
        files.extend(page.get('Contents', []))
    if not files:
        return None
    sizes = [f['Size'] for f in files]
    return {
        'file_count': len(sizes),
        'total_kb': sum(sizes) / 1024,
        'avg_kb': sum(sizes) / len(sizes) / 1024,
        'min_kb': min(sizes) / 1024
    }

firehose_stats = get_prefix_stats(f'streaming/sequencer_runs/')
direct_stats = get_prefix_stats(f'streaming/direct_json/')

comparison = pd.DataFrame([
    {'approach': 'Direct JSON to S3', **(direct_stats or {'file_count': len(all_sent_events), 'avg_kb': 0.25})},
    {'approach': 'Firehose → Parquet', **(firehose_stats or {'file_count': 1, 'avg_kb': 'pending'})}
]).set_index('approach')

print("Direct JSON vs Firehose Parquet comparison:")
print(comparison.to_string())
print()
print("Rule: Athena opens each file with overhead of ~10ms.")
print(f"Direct approach ({len(all_sent_events)} files): ~{len(all_sent_events)*10}ms overhead just to open files.")
print("Firehose approach (1-2 files): ~20ms overhead.")
````

## Section 8: Cleanup

````python
def cleanup_streaming():
    print("Cleaning up streaming resources...")
    try:
        firehose.delete_delivery_stream(DeliveryStreamName=FIREHOSE_STREAM, AllowForceDelete=True)
        print(f"  Deleted Firehose stream: {FIREHOSE_STREAM}")
    except Exception as e:
        print(f"  Firehose deletion: {e}")
    try:
        iam.delete_role_policy(RoleName=FIREHOSE_ROLE, PolicyName='FirehoseS3GlueAccess')
        iam.delete_role(RoleName=FIREHOSE_ROLE)
        print(f"  Deleted IAM role: {FIREHOSE_ROLE}")
    except Exception as e:
        print(f"  IAM cleanup: {e}")
    print("Done.")

print("Run cleanup_streaming() when finished with the exercise.")
````
