# Example 01: Hello World - Your First AI-Generated Code

## Learning Objective
Learn how to prompt Claude Code to generate simple programs and understand the basics of AI-assisted coding.

---

## The Prompt

Try asking Claude Code:

```
Write a Python program that prints "Hello, World!" and then asks for the user's name and greets them personally.
```

## Expected Output

Claude Code will generate something like this:

````python
def main():
    print("Hello, World!")
    
    name = input("What is your name? ")
    print(f"Nice to meet you, {name}!")

if __name__ == "__main__":
    main()
````

## Key Concepts Demonstrated

1. **Clear Instructions**: The prompt clearly states what the program should do
2. **Step-by-Step Requirements**: Two distinct actions are requested
3. **Language Specification**: We specified "Python" to get Python code

## What to Notice

- Claude Code provides clean, well-structured code
- The code follows Python best practices (using `if __name__ == "__main__"`)
- f-strings are used for modern string formatting
- The code is wrapped in a `main()` function for better organization

## Common Mistakes to Avoid

1. **Being too vague**: "Write a hello world" is less clear than specifying the language and exact behavior
2. **Overcomplicating**: Start simple and iterate
3. **Not testing**: Always run the generated code to verify it works
